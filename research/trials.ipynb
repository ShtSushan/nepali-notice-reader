```python
import sys
sys.path.append('..')

from backend.services.gemini_service import extract_text_from_image
from backend.services.groq_service import translate_to_english, summarize_notice
from backend.services.embed_service import process_text_for_storage

# Step 1 — OCR
nepali_text = extract_text_from_image(r"C:\Users\shtsu\OneDrive\Desktop\nepali_notice_reader\research\Bachelor-in-Information-Technology-(BIT)-Admission-Entrance-Notice-from-TUIOST.png")

# Step 2 — Translate
english_text = translate_to_english(nepali_text)

# Step 3 — Summarize
summary = summarize_notice(english_text)

# Step 4 — Chunk + Embed
chunks = process_text_for_storage(english_text)

print(f"Total chunks: {len(chunks)}")
print(f"\nFirst chunk text:\n{chunks[0]['chunk_text']}")
print(f"\nEmbedding dimension: {len(chunks[0]['embedding'])}")
print(f"First 5 values: {chunks[0]['embedding'][:5]}")
print(f"\nSummary: {summary}")
```

```python
from backend.services.groq_service import answer_question
from backend.services.embed_service import embed_text
from backend.services.db_service import search_similar_chunks, save_message, get_chat_history

question = "What is the application fee?"

# Step 1 — embed question
question_vector = embed_text(question)

# Step 2 — find relevant chunks
chunks = search_similar_chunks(notice["id"], question_vector, top_k=3)

# Step 3 — get chat history
history = get_chat_history(notice["id"])

# Step 4 — get answer from Groq
answer = answer_question(question, chunks, history)
print("Answer:", answer)

# Step 5 — save Q&A to DB
save_message(notice["id"], "user", question)
save_message(notice["id"], "assistant", answer)
print("Conversation saved ✅")
```
